## Scaling and Encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_selector
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

preprocessor = ColumnTransformer(
    transformers=[
        ("num", MinMaxScaler(), make_column_selector(dtype_include="float64")),
        ("cat", OneHotEncoder(), make_column_selector(dtype_include="object")),
    ],
    remainder='passthrough'
)
```

```python
X = df.drop(['is_fraud'], axis = 1)
X
```

```python
y = df[['is_fraud']]
y
```

```python
# DO NOT waste resources to run this unless the shape is changed
preprocessor.fit_transform(X).shape
```

## Create Model

```python
from tensorflow import keras

def create_model():
    model = keras.models.Sequential([
        keras.Input(shape=(19,)),
        keras.layers.Dense(12, activation="relu"),
        keras.layers.Dense(6, activation="relu"),
        keras.layers.Dense(1, activation="sigmoid")
    ])

    model.compile(
        optimizer=keras.optimizers.SGD(learning_rate=0.1), 
        loss=keras.losses.BinaryCrossentropy(), 
        metrics=[
            "accuracy", 
            keras.metrics.Precision(name="precision"),
            keras.metrics.Recall(name="recall")
        ]
    )

    return model
```

```python
from sklearn.pipeline import Pipeline
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

clf = Pipeline(
    steps=[
        ("preprocessor", preprocessor), 
        ("classifier", KerasClassifier(build_fn=create_model))
    ]
)
```

```python
from sklearn import set_config

set_config(display="diagram")
clf
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "classifier__epochs": [16],
}

grid_search = GridSearchCV(clf, param_grid, cv=10)
grid_search
```
